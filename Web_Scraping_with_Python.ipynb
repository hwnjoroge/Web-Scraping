## <font color='#2F4F4F'>Prerequisites</font>

```
# We first import the required libraries
# ---
#
import pandas as pd             # library for data manupation
import requests                 # library for fetching a web page 
from bs4 import BeautifulSoup   # library for extrating contents from a webpage 
```

## <font color='#2F4F4F'>Step 1: Obtaining our Data</font>

```
# PigiaMe: https://www.pigiame.co.ke/it-software-jobs
# ---
#store website in variable and get request and ststus code

pigia_me = requests.get('https://www.pigiame.co.ke/it-software-jobs')
pigia_me
```

```
# MyJobMag: https://www.myjobmag.co.ke/jobs-by-field/information-technology

#store website in variable and get request and status code

my_job_mag = requests.get('https://www.myjobmag.co.ke/jobs-by-field/information-technology')
my_job_mag
```

```
# KenyanJob: https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133

#store website in variable and get request and status code

kenyan_job = requests.get('https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133')
kenyan_job
```

## <font color='#2F4F4F'>Step 2: Parsing</font>

```
# Parsing our document: pigia_me
pigia_me = pigia_me.text
```

```
#Creating a soup object
# ---
pigia_me_soup = BeautifulSoup(pigia_me, 'html.parser')
pigia_me_soup
```

```
# Parsing our document: my_job_mag
my_job_mag = my_job_mag.text
```

```
#Creating a soup object
# ---
my_job_mag_soup = BeautifulSoup(my_job_mag, 'html.parser')
my_job_mag_soup
```

```
# Parsing our document: kenyan_job
kenyan_job = kenyan_job.text
```

```
# Parsing our document: kenyan_job
# ---
kenyan_job_soup = BeautifulSoup(kenyan_job, 'html.parser')
kenyan_job_soup
```

## <font color='#2F4F4F'>Step 3: Extracting Required Elements</font>

### <font color='#2F4F4F'>Pigia Me</font>

```
results_pigia2 = pigia_me_soup.find_all('div',{'class':'listing-card listing-card--tab listing-card--has-content listing-card--highlight-placeholder'})
```

```
results_pigia2
```

```
# 2a. Extracting job titles: pigia me
results_pigia2[2].find('div',{'class':'listing-card__header__title'}).get_text().strip().replace('\n',', ')
```

```
# 2b. Extracting job links: pigia me
link = results_pigia2[2].find('a').get('href')
print(link)
```

```
results_pigia2[0].find('a').get('href')
```

#### Put everything together inside a For-Loop

```
#Pigia Me Job Titles :  looping through the results_pigia2 for the job title.

job_title= []  #intitialize the job_title
for i in range (1,2):

 for result in results_pigia2:

    try:
        job_title.append(result.find('div',{'class':'listing-card__header__title'}).get_text().strip().replace('\n',', '))
    except:
        job_title.append('n/a')
```

```
#Pigia Me Job Links :  looping through the results_pigia2 for the job links.

Job_link = [] #initializing the job_link 

for i in range (1,2):

 for result in results_pigia2:
  
    try:
        Job_link.append(result.find('a').get('href'))
    except:
        Job_link.append('n/a')
```

#### Create a Pandas Dataframe

```
PigiaMe_df = pd.DataFrame({'Job Title': job_title , 'Link': Job_link })
```

```
PigiaMe_df
```

### <font color='#2F4F4F'>My_job_mag</font>

```
results = my_job_mag_soup.find_all('h2')
results
```

```
len(results)
```

```
# 2a. Extracting job titles: my_job_mag
results[1].find('a').get_text()
```

```
# 2b. Extracting job links: my_job_mag
#Extracting the relative url from results and assigning a variable

relative_url = results[0].find('a').get('href')
relative_url
```

```
results[0].find('a').get('href')
```

```
#root url
root_url = 'https://www.myjobmag.co.ke'
```

```
#combine URLs

#combining the root url and the relative url
links_my_job_mag = root_url + relative_url
links_my_job_mag
```

#### Put everything together inside a For-Loop

```
#Job  Titles looping through the results for the job title.

job_title= []  #intitialize the job_title
for i in range (1,2):

 for result in results:

    try:
        job_title.append(result.find('a').get_text())
    except:
        job_title.append('n/a')
```

```
#MyJobMag Job Links:  looping through the results for the job links .

relative_url = [] #initializing the job_link 

for i in range (1,2):

 for result in results:
  
    try:
        relative_url.append(result.find('a').get('href')) 
    except:
        relative_url.append('n/a')
```

```
# Looping and combining the urls from the loop with the root url

import urllib.parse 
url_combined = []

for link in relative_url:
    url_combined.append(urllib.parse.urljoin(root_url, link))
```

#### Create a Pandas Dataframe

```
# pandas dataframe takes a dictionary
# dictionary: key-value pair
MyJobMag_df= pd.DataFrame({ 'Job Title': job_title, 'Link': url_combined })
```
